# Beauti notebook

> Directives are signals to render, but how to make notebook prettier?

All method to make your notebook beautiful is already noted in [Quarto authoring](https://quarto.org/docs/authoring/markdown-basics.html#overview). This tutorial help you nagivate through the linked page and point out additional cases.

*Just access the link, follow the instruction!*

:::{.callout-caution}
It is typical to have an unsatisfied result when execute in Jupyter since it use a different set of render engine. Try `nbdev_preview` for better interaction. 
:::
